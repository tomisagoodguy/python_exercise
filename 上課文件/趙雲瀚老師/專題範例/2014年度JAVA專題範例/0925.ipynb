```python
import matplotlib.pyplot as plt
import networkx as nx
from matplotlib.patches import Rectangle
import matplotlib.font_manager as fm

# 嘗試多個中文字體
chinese_fonts = ['DFKai-SB', 'Microsoft YaHei',
                 'SimHei', 'SimSun', 'NSimSun', 'FangSong', 'KaiTi']

# 找到可用的中文字體
chinese_font = None
for font in chinese_fonts:
    if any([f for f in fm.fontManager.ttflist if font.lower() in f.name.lower()]):
        chinese_font = font
        break

if chinese_font:
    plt.rcParams['font.family'] = chinese_font
else:
    print("未找到合適的中文字體，將使用默認字體。")

# 增加全局字體大小和行距
plt.rcParams['font.size'] = 12
plt.rcParams['lines.linewidth'] = 2


def draw_waterfall_diagram():
    fig, ax = plt.subplots(figsize=(16, 8))
    ax.set_title("CIM-2 水道圖", fontsize=16)

    steps = ["會員", "WEB介面", "應用程式層", "資料庫層"]
    y_positions = [4, 3, 2, 1]

    for i, (step, y) in enumerate(zip(steps, y_positions)):
        ax.text(-0.5, y, step, ha='right', va='center', fontsize=14)
        ax.axhline(y=y, color='black', linestyle='--', xmin=0.05, xmax=0.95)

    ax.set_ylim(0.5, 4.5)
    ax.set_xlim(-1, 6)
    ax.axis('off')

    ax.annotate("", xy=(1, 3), xytext=(1, 4),
                arrowprops=dict(arrowstyle="->", lw=2))
    ax.annotate("", xy=(2, 2), xytext=(2, 3),
                arrowprops=dict(arrowstyle="->", lw=2))
    ax.annotate("", xy=(3, 1), xytext=(3, 2),
                arrowprops=dict(arrowstyle="->", lw=2))

    ax.annotate("", xy=(4, 2), xytext=(4, 1),
                arrowprops=dict(arrowstyle="->", lw=2))
    ax.annotate("", xy=(5, 3), xytext=(5, 2),
                arrowprops=dict(arrowstyle="->", lw=2))

    plt.tight_layout()
    plt.show()


def draw_use_case_diagram():
    fig, ax = plt.subplots(figsize=(14, 10))
    ax.set_title("CIM-3A 系統UC圖", fontsize=16)

    rect = Rectangle((0.2, 0.1), 0.6, 0.8, fill=False)
    ax.add_patch(rect)
    ax.text(0.5, 0.95, "訂位訂餐系統", ha='center', fontsize=14)

    ax.text(0.1, 0.7, "會員", ha='center', fontsize=14)
    ax.text(0.1, 0.3, "客服人員", ha='center', fontsize=14)

    use_cases = [
        (0.5, 0.8, "訂位"),
        (0.5, 0.65, "訂餐"),
        (0.5, 0.5, "查詢"),
        (0.5, 0.35, "取消訂位")
    ]

    for x, y, label in use_cases:
        ax.add_patch(plt.Circle((x, y), 0.05, fill=False))
        ax.text(x, y, label, ha='center', va='center', fontsize=12)

    ax.plot([0.15, 0.45], [0.7, 0.8], 'k-', lw=2)
    ax.plot([0.15, 0.45], [0.7, 0.65], 'k-', lw=2)
    ax.plot([0.15, 0.45], [0.7, 0.5], 'k-', lw=2)
    ax.plot([0.15, 0.45], [0.3, 0.35], 'k-', lw=2)

    ax.axis('off')
    plt.tight_layout()
    plt.show()


def draw_activity_diagram():
    fig, ax = plt.subplots(figsize=(14, 16))
    ax.set_title("CIM-3B 活動圖", fontsize=16)

    activities = [
        (0.5, 0.95, "開始"),
        (0.5, 0.85, "登入系統"),
        (0.5, 0.75, "選擇功能"),
        (0.3, 0.65, "訂位"),
        (0.7, 0.65, "查詢"),
        (0.3, 0.55, "訂餐"),
        (0.7, 0.55, "取消訂位"),
        (0.5, 0.45, "確認操作"),
        (0.5, 0.35, "更新資料庫"),
        (0.5, 0.25, "顯示結果"),
        (0.5, 0.15, "結束")
    ]

    for x, y, label in activities:
        if label in ["開始", "結束"]:
            ax.add_patch(plt.Circle((x, y), 0.03, fill=True))
        else:
            ax.add_patch(Rectangle((x-0.1, y-0.03), 0.2, 0.06, fill=False))
        ax.text(x, y, label, ha='center', va='center', fontsize=12)

    ax.annotate("", xy=(0.5, 0.82), xytext=(0.5, 0.92),
                arrowprops=dict(arrowstyle="->", lw=2))
    ax.annotate("", xy=(0.5, 0.72), xytext=(0.5, 0.82),
                arrowprops=dict(arrowstyle="->", lw=2))
    ax.annotate("", xy=(0.3, 0.68), xytext=(0.5, 0.72),
                arrowprops=dict(arrowstyle="->", lw=2))
    ax.annotate("", xy=(0.7, 0.68), xytext=(0.5, 0.72),
                arrowprops=dict(arrowstyle="->", lw=2))
    ax.annotate("", xy=(0.3, 0.58), xytext=(0.3, 0.62),
                arrowprops=dict(arrowstyle="->", lw=2))
    ax.annotate("", xy=(0.7, 0.58), xytext=(0.7, 0.62),
                arrowprops=dict(arrowstyle="->", lw=2))
    ax.annotate("", xy=(0.5, 0.48), xytext=(0.3, 0.52),
                arrowprops=dict(arrowstyle="->", lw=2))
    ax.annotate("", xy=(0.5, 0.48), xytext=(0.7, 0.52),
                arrowprops=dict(arrowstyle="->", lw=2))
    ax.annotate("", xy=(0.5, 0.38), xytext=(0.5, 0.42),
                arrowprops=dict(arrowstyle="->", lw=2))
    ax.annotate("", xy=(0.5, 0.28), xytext=(0.5, 0.32),
                arrowprops=dict(arrowstyle="->", lw=2))
    ax.annotate("", xy=(0.5, 0.18), xytext=(0.5, 0.22),
                arrowprops=dict(arrowstyle="->", lw=2))

    ax.axis('off')
    plt.tight_layout()
    plt.show()


def draw_system_architecture():
    G = nx.DiGraph()
    G.add_edge("訂位訂餐系統", "會員管理")
    G.add_edge("訂位訂餐系統", "訂位管理")
    G.add_edge("訂位訂餐系統", "訂餐管理")
    G.add_edge("訂位訂餐系統", "查詢功能")

    G.add_edge("會員管理", "會員資料維護")
    G.add_edge("會員管理", "會員驗證")

    G.add_edge("訂位管理", "新增訂位")
    G.add_edge("訂位管理", "修改訂位")
    G.add_edge("訂位管理", "取消訂位")

    G.add_edge("訂餐管理", "選擇菜單")
    G.add_edge("訂餐管理", "確認訂單")

    G.add_edge("查詢功能", "訂位查詢")
    G.add_edge("查詢功能", "訂餐查詢")

    fig, ax = plt.subplots(figsize=(16, 12))
    pos = nx.spring_layout(G, k=0.9, iterations=50)
    nx.draw(G, pos, with_labels=True, node_color='lightblue',
            node_size=4000, font_size=10, font_weight='bold',
            arrows=True, ax=ax)

    ax.set_title("CIM-Milestone: 系統功能架構圖", fontsize=16)
    plt.tight_layout()
    plt.show()


def draw_robustness_diagram():
    fig, ax = plt.subplots(figsize=(16, 12))
    ax.set_title("PIM-1 強韌分析圖", fontsize=16)

    boundary_objects = [
        (0.2, 0.8, "登入介面"),
        (0.2, 0.6, "訂位介面"),
        (0.2, 0.4, "訂餐介面"),
        (0.2, 0.2, "查詢介面")
    ]

    for x, y, label in boundary_objects:
        ax.add_patch(plt.Circle((x, y), 0.05, fill=False))
        ax.text(x, y-0.08, label, ha='center', fontsize=12)

    control_objects = [
        (0.5, 0.8, "驗證控制"),
        (0.5, 0.6, "訂位控制"),
        (0.5, 0.4, "訂餐控制"),
        (0.5, 0.2, "查詢控制")
    ]

    for x, y, label in control_objects:
        ax.add_patch(plt.Rectangle((x-0.05, y-0.05), 0.1, 0.1, fill=False))
        ax.text(x, y-0.08, label, ha='center', fontsize=12)

    entity_objects = [
        (0.8, 0.8, "會員資料"),
        (0.8, 0.6, "訂位資料"),
        (0.8, 0.4, "訂餐資料"),
        (0.8, 0.2, "菜單資料")
    ]

    for x, y, label in entity_objects:
        ax.add_patch(plt.Rectangle((x-0.08, y-0.03), 0.16, 0.06, fill=False))
        ax.text(x, y-0.08, label, ha='center', fontsize=12)

    for i in range(4):
        ax.plot([0.25, 0.45], [0.8-i*0.2, 0.8-i*0.2], 'k-', lw=2)
        ax.plot([0.55, 0.72], [0.8-i*0.2, 0.8-i*0.2], 'k-', lw=2)

    ax.text(0.02, 0.9, "會員", ha='left', fontsize=14)
    ax.plot([0.05, 0.15], [0.8, 0.8], 'k-', lw=2)
    ax.plot([0.05, 0.15], [0.6, 0.6], 'k-', lw=2)
    ax.plot([0.05, 0.15], [0.4, 0.4], 'k-', lw=2)
    ax.plot([0.05, 0.15], [0.2, 0.2], 'k-', lw=2)

    ax.text(0.02, 0.1, "客服人員", ha='left', fontsize=14)
    ax.plot([0.05, 0.15], [0.2, 0.2], 'k-', lw=2)

    ax.axis('off')
    plt.tight_layout()
    plt.show()


# 執行所有圖表繪製函數
draw_waterfall_diagram()
draw_use_case_diagram()
draw_activity_diagram()
draw_system_architecture()
draw_robustness_diagram()
```
